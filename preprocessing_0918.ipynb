```python
import re
```

```python
def regex(text):
    text = re.sub("([a-zA-Z0-9_.+-]+@[a-zA-Z0-9-]+\.[a-zA-Z-.]+)","", text)
    text = re.sub("[\r\n\xa0]", "", text)
    text = re.sub("(\.\s+[ㄱ-ㅎ가-힣]+\s[기]+[자]+)", "", text)
    text = re.sub("[^\w\s^.]", " ", text)
    text = re.sub("[0-9]", '', text)
    return text                                                        #보겸님>형석님이 주신 정규표현식 코드 사용
```

```python
#불용어 처리 https://github.com/jaepil-choi/asiae-BOK-nlp/blob/master/03_preprocessing_text_data(old).ipynb 참고

def getStopwords(data_path) :
    f = open(data_path, 'r',-1,'utf-8')
    stop_word = []

    while True:
        line = f.readline()
        stop_word.append(line.strip())
        if not line: break

    return stop_word
```
