```python
from tqdm import tqdm
import time
```

```python
for index, row in tqdm(comments_df.iterrows()):
  time.sleep(0.25)
  target = row["comment_date"]
  if "T" in target:
    target_TMD,target_time = target.split("T")
    if "+0900" in  target_time:
      new = target_TMD + "_" + target_time[:-6]
      comments_df.loc[index,"comment_date"] = new
    else:
      if target_time[-1] == "\n":
        target_time = target_time[:-1]
      new = target_TMD + "_" + target_time
      comments_df.loc[index,"comment_date"] = new

```
