```python
def extract_journalist_names(text):
    pattern = re.compile(r'(?:(?:워싱턴|대전)=)?([가-힣\s]+)(?:\s*(?:특파원|기자))?')
    matches = pattern.findall(text)
    with open('stopwords.txt', 'r') as f:
        stop_words = f.read().split("\n")
    journalist_names = []
    for match in matches:
        name = re.sub(r'(?:워싱턴|대전)=|(?:특파원)', '', match).strip()
        for remove in stop_words:
            name = name.replace(remove,'')
        journalist_names.append(name)


    return journalist_names

article_df["reporter"] = article_df["reporter"].apply(lambda x: ','.join(extract_journalist_names(x)))

article_df
```
